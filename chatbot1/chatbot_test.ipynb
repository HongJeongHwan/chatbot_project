```python
# 필요한 라이브러리 임포트
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer
from tqdm import tqdm
tqdm.pandas() 
```

```python
# 문장 임베딩 생성용 데이터 불러오기

# 파일경로 설정
file_path = 'Data/ChatBotData.csv'

# pd.read_csv() 사용 --> DataFrame 생성
df = pd.read_csv(file_path)

# 결과확인
print(df)
```

```python
# 사전학습된 한국어 SentenceBERT 모델 생성
model_name = 'ddobokki/klue-roberta-base-nli-sts'
model = SentenceTransformer(model_name)
```

```python
# 모델 테스트 : 텍스트 --> 임베딩

# 테스트 문장 생성
sentence = "본 발명은 바위수염 추출물을 포함하는 항비만용 조성물 및 그 제조방법에 관한 것이다"

# model.encode(sentence) 사용
embedding = model.encode(sentence)

# 문장 임베딩결과 확인
print(f'임베딩 벡터의 모양 확인 : {embedding.shape}')
print(f'임베딩 벡터 확인 :\n{embedding}')
```

```python
# 문장임베딩 생성
'''
apply(lambda x: model.encode(x)) 적용
'''
# 임베딩 벡터 생성
embeddings = df.loc[:,'Q'].apply(lambda x: model.encode(x)).values

# 결과확인
print(f'문장임베딩벡터  전체의 모양 : {embeddings.shape}')
print(f'문장임베딩벡터  전체 확인 :\n{embeddings}')
```

```python
df_chat = pd.read_csv('Data/ChatBotEmbeddings.csv')
```

```python
# 두 문장의 텍스트유사도 계산 함수 정의

# 필요한 함수 임포트
from numpy import dot # 벡터의 내적을 구하기
from numpy.linalg import norm

# 사용자정의함수
def cos_sim(A,B):
    return dot(A,B) / (norm(A)*norm(B))
```

```python
# 사용자 질문에 대한 응답 구현
'''
1. 사용자의 질문이 들어오면 해당 질문에 대한 문장 임베딩 벡터를 구한다.
2. 사용자 질문의 문장 임베딩 값과 챗봇 데이터의 임베딩 컬럼(df.loc[:, 'embedding'])에
저장해둔 모든 질문 샘플들의 문장 임베딩 값들을 이용, 코사인 유사도를 계산
--> 코사인 유사도 값을 저장하는 컬럼(df.loc[:, 'score'])을 생성하고 값을 저장
3. 코사인 유사도 값이 가장 높은 질문 샘플을 찾는다.
4. 해당 질문 샘플과 짝이 되는 답변 샘플을 반환한다.
'''
df = df_chat
# 사용자정의함수 생성
def chatbot(question):

    # 사용자 질문 문장 --> 문장 임베딩 벡터 구하기
    embedding = model.encode(question)

    # 코사인유사도 계산 --> 코사인유사도 값을 컬럼(score)으로 저장
    df['score'] = df.loc[:,'embeddings'].apply(lambda x: cos_sim(x,embedding)).values

    # 코사인유사도 값이 가장 큰 질문 샘플을 찾아서 해당질문과 짝이되는 답변 추출
    # score = df['score'].sort_values(ascending=False).iloc[0]
    # cond = df.loc[:,'score'].values == score
    # result = df.loc[cond,'A'].values[0]
    result = df.loc[df['score'].idxmax(),'A']

    return result
```

```python
# 챗봇성능 테스트(1)

# 사용자 입력
text = input()

# 챗봇의 답변
ans = chatbot(text)
print(ans)
# print(chatbot(input()))
```
